```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

2. Загрузить данные в датафрейм

```python
df = pd.read_csv('google_review_ratings.csv')
```

3. Вывести статистическую информацию о наборе данных

```python
print(df.describe())
```

4. Вывести названия столбцов и строк

```python
print("Названия столбцов:", df.columns)
print("Количество строк:", len(df))
```

5. Заменить категориальные данные количественными

```python
df_encoded = pd.get_dummies(df)
```

6. Визуализировать данные с помощью Matplotlib и Seaborn. Должно быть не менее 2х визуализаций на каждый признак.

```python
for column in df.select_dtypes(include=['float64', 'int64']).columns:
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    sns.histplot(df[column], kde=True)
    plt.title(f'Гистограмма для признака: {column}')
    plt.xlabel(column)
    plt.ylabel('Частота')

    plt.subplot(1, 2, 2)
    sns.scatterplot(x=df.index, y=df[column])
    plt.title(f'Диаграмма рассеяния для признака: {column}')
    plt.xlabel('Индекс')
    plt.ylabel(column)

    plt.tight_layout()
    plt.show()
```

7. Должно быть не менее 3х групповых визуализаций

```python
group_column = 'Category 2'

plt.figure(figsize=(10, 6))
sns.boxplot(x=group_column, y=df['Category 3'], data=df)
plt.title(f'Boxplot для рейтинга курортов и {group_column}')
plt.xlabel(group_column)
plt.ylabel('Значение Category 3')
plt.show()

plt.figure(figsize=(10, 6))
sns.barplot(x=group_column, y=df['Category 4'], data=df, errorbar=None)
plt.title(f'Barplot для рейтинга пляжей и {group_column}')
plt.xlabel(group_column)
plt.ylabel('Среднее значение Category 4')
plt.show()

plt.figure(figsize=(10, 6))
sns.violinplot(x=group_column, y=df['Category 5'], data=df)
plt.title(f'Violin plot для рейтинга парков и {group_column}')
plt.xlabel(group_column)
plt.ylabel('Значение Category 5')
plt.show()
```
